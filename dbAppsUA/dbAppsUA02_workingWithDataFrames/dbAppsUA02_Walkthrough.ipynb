Ми працюватимемо з набором даних **Titanic** — реальними історичними даними пасажирів Титаніка.

## Розділ 1: Завантаження даних

```python
# Імпортуємо бібліотеку pandas
import pandas as pd

# Завантажуємо набір даних Titanic у DataFrame
titanic = pd.read_csv("Titanic Dataset.csv")

# Швидка перевірка — розмір даних
print("Розмір:", titanic.shape)

# Перші 5 рядків
titanic.head()
```

## Розділ 2: Вибір стовпців (Selecting Columns)

```python
# Вибираємо ОДИН стовпець — повертає Series (один стовпець даних)
# Це як SQL: SELECT name FROM titanic
names = titanic["name"]

print("Перші 5 імен:")
print(names.head())
print()
print(f"Тип даних: {type(names)}")
```

```python
# Вибираємо КІЛЬКА стовпців — використовуємо ПОДВІЙНІ дужки
# Це як SQL: SELECT name, age, survived FROM titanic
subset = titanic[["name", "age", "survived"]]

print("Перші 5 рядків вибраних стовпців:")
print(subset.head())
```

## Розділ 3: Фільтрація рядків (Filtering Rows)

```python
# Фільтруємо: залишаємо тільки рядки де умова True
# pclass == 1 (пасажири першого класу)
# Це як SQL: SELECT * FROM titanic WHERE pclass = 1
firstClass = titanic[titanic["pclass"] == 1]

print(f"Пасажирів першого класу: {len(firstClass)}")
print()
firstClass[["name", "pclass", "fare"]].head()
```

```python
# Різні оператори порівняння

# Пасажири старші за 50
olderPassengers = titanic[titanic["age"] > 50]
print(f"Пасажирів старших за 50: {len(olderPassengers)}")

# Пасажири які вижили (survived == 1)
survivors = titanic[titanic["survived"] == 1]
print(f"Пасажирів які вижили: {len(survivors)}")
```

```python
# Комбінуємо КІЛЬКА умов з & (і) або | (або)
# ВАЖЛИВО: Кожна умова потребує своїх дужок!

# Жінки які вижили — ОБІ умови мають бути True
# Це як SQL: WHERE sex = 'female' AND survived = 1
femaleSurvivors = titanic[
    (titanic["sex"] == "female") & (titanic["survived"] == 1)
]
print(f"Жінок які вижили: {len(femaleSurvivors)}")

# Пасажири першого АБО другого класу
upperClass = titanic[
    (titanic["pclass"] == 1) | (titanic["pclass"] == 2)
]
print(f"Пасажирів вищих класів: {len(upperClass)}")
```

## Розділ 4: Сортування даних (Sorting)

```python
# Сортуємо за віком — за зростанням (ascending) за замовчуванням
# Це як SQL: SELECT * FROM titanic ORDER BY age
byAge = titanic.sort_values("age")

print("10 наймолодших пасажирів:")
byAge[["name", "age"]].head(10)
```

```python
# Сортуємо за СПАДАННЯМ — від найбільшого до найменшого
# Це як SQL: SELECT * FROM titanic ORDER BY fare DESC
byFare = titanic.sort_values("fare", ascending=False)

print("10 найдорожчих квитків:")
byFare[["name", "fare", "pclass"]].head(10)
```

## Розділ 5: Базова статистика (Basic Statistics)

```python
# Обчислюємо статистику для числових стовпців
avgAge = titanic["age"].mean()              # Середній вік
maxFare = titanic["fare"].max()             # Найвища ціна квитка
minAge = titanic["age"].min()               # Наймолодший пасажир
totalSurvived = titanic["survived"].sum()   # Загальна кількість тих хто вижив

print(f"Середній вік: {avgAge:.1f} років")
print(f"Найвища ціна квитка: ${maxFare:.2f}")
print(f"Наймолодший пасажир: {minAge} років")
print(f"Всього вижило: {int(totalSurvived)}")
```

```python
# Спочатку фільтруємо, потім рахуємо — дуже потужний підхід!

# Середня ціна квитка для першого класу
avgFirstFare = titanic[titanic["pclass"] == 1]["fare"].mean()

# Середня ціна квитка для третього класу
avgThirdFare = titanic[titanic["pclass"] == 3]["fare"].mean()

print(f"Середня ціна першого класу: ${avgFirstFare:.2f}")
print(f"Середня ціна третього класу: ${avgThirdFare:.2f}")
print(f"Різниця: ${avgFirstFare - avgThirdFare:.2f}")
```

## Розділ 6: Підрахунок значень (Value Counts)

```python
# .value_counts() рахує скільки разів з'являється кожне значення
# Це як SQL: SELECT pclass, COUNT(*) FROM titanic GROUP BY pclass

print("Пасажири за класом:")
print(titanic["pclass"].value_counts())
print()

print("Виживання (0 = не вижив, 1 = вижив):")
print(titanic["survived"].value_counts())
print()

print("Розподіл за статтю:")
print(titanic["sex"].value_counts())
```

## Підсумок — Міст від Pandas до SQL

| Завдання | Pandas | SQL |
|----------|--------|-----|
| Вибір стовпців | `df[["col1", "col2"]]` | `SELECT col1, col2 FROM table` |
| Фільтрація рядків | `df[df["col"] == value]` | `WHERE col = value` |
| Сортування | `df.sort_values("col")` | `ORDER BY col` |
| Підрахунок | `len(df)` | `COUNT(*)` |
| Середнє | `df["col"].mean()` | `AVG(col)` |
| Групування | `df["col"].value_counts()` | `GROUP BY col` |

**Далі:** У наступному уроці ми завантажимо ці дані в **справжню базу даних SQLite**!
