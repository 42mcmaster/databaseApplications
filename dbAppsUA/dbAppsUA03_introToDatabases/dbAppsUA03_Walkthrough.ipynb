## Розділ 1: Що таке база даних?

**Дані (Data)** — це необроблені факти (наприклад, "25", "жінка", "так")

**Інформація (Information)** — це оброблені дані з контекстом (наприклад, "25-річна жінка-пасажирка вижила")

**База даних (Database)** — організована колекція пов'язаних даних. Переваги над CSV:
- **Структура**: Таблиці з визначеними стовпцями та типами даних
- **Зв'язки**: Таблиці можуть посилатися одна на одну
- **Ефективність**: Швидкі запити на великих наборах даних
- **Цілісність**: Правила для якості даних

**SQLite** — легка файлова система баз даних, вбудована в Python.

## Розділ 2: Налаштування та імпорт

```python
# Імпортуємо потрібні бібліотеки
import pandas as pd     # Для DataFrame (ви це вже знаєте)
import sqlite3          # Для роботи з базами даних SQLite (вбудовано в Python)

print("Бібліотеки імпортовано успішно!")
```

## Розділ 3: Завантаження CSV (повторення)

```python
# Завантажуємо CSV Titanic у DataFrame — як в Уроках 01-02
titanic = pd.read_csv("Titanic Dataset.csv")

print(f"Завантажено {titanic.shape[0]} рядків і {titanic.shape[1]} стовпців")
titanic.head()
```

## Розділ 4: Створення бази даних SQLite

**`sqlite3.connect()`** — це ваш вхід до SQLite:
- Створює новий файл `.db` (якщо не існує) або відкриває існуючий
- Повертає **об'єкт з'єднання (connection)** — ваш зв'язок з базою даних

```python
# Створюємо (або відкриваємо) файл бази даних SQLite
conn = sqlite3.connect("titanic.db")

print("Підключено до titanic.db!")
```

## Розділ 5: Збереження DataFrame у базу даних

**`to_sql()`** — метод pandas, який з'єднує DataFrame з базою даних:
- Перший аргумент: назва таблиці
- `if_exists="replace"` — перезаписати якщо існує
- `index=False` — не зберігати номери рядків

```python
# Зберігаємо DataFrame як таблицю "passengers" у базі даних
titanic.to_sql("passengers", conn, if_exists="replace", index=False)

print("Дані збережено в таблицю 'passengers' у titanic.db!")
```

## Розділ 6: Ваш перший SQL запит

**`pd.read_sql()`** виконує SQL запит і повертає результат як DataFrame:
- Перший аргумент: SQL рядок (ваш запит)
- Другий аргумент: з'єднання з базою даних
- Повертає: DataFrame з результатами

```python
# Виконуємо SQL запит і отримуємо результат як DataFrame
# SELECT * FROM passengers — "отримати всі стовпці з таблиці passengers"
# LIMIT 5 — "тільки перші 5 рядків"
result = pd.read_sql("SELECT * FROM passengers LIMIT 5", conn)
result
```

```python
# Вибираємо конкретні стовпці — як titanic[["name", "age"]] у pandas
result = pd.read_sql("SELECT name, age, survived FROM passengers LIMIT 10", conn)
result
```

## Розділ 7: Дослідження структури бази даних

```python
# Список всіх таблиць у базі даних
tables = pd.read_sql("SELECT name FROM sqlite_master WHERE type='table'", conn)
print("Таблиці в базі даних:")
print(tables)
```

```python
# Інформація про стовпці таблиці passengers
# PRAGMA table_info() — команда SQLite для деталей стовпців
columnInfo = pd.read_sql("PRAGMA table_info(passengers)", conn)
columnInfo
```

## Розділ 8: Порівняння Pandas і SQL

Ви вже знаєте як це робити в pandas. Подивімось SQL еквівалент.

```python
# Спосіб PANDAS (з Уроку 02):
pandasResult = titanic[["name", "age"]].head(5)
print("Результат Pandas:")
print(pandasResult)

print("\n" + "="*50 + "\n")

# Спосіб SQL (новий!):
sqlResult = pd.read_sql("SELECT name, age FROM passengers LIMIT 5", conn)
print("Результат SQL:")
print(sqlResult)

print("\nТі самі дані, інший підхід!")
```

## Розділ 9: Закриття з'єднання

Завжди закривайте з'єднання коли закінчили роботу!

```python
# Закриваємо з'єднання з базою даних
conn.close()

print("З'єднання закрито. Базу даних збережено!")
```

## Підсумок

| Завдання | Код | Призначення |
|----------|-----|-------------|
| Імпорт SQLite | `import sqlite3` | Додати функціонал БД |
| Підключення | `conn = sqlite3.connect("file.db")` | Відкрити або створити БД |
| Зберегти DataFrame | `df.to_sql("table", conn, ...)` | Створити таблицю з DataFrame |
| Виконати запит | `pd.read_sql("SELECT ...", conn)` | SQL → DataFrame |
| Закрити | `conn.close()` | Зберегти і закрити БД |
