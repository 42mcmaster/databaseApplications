## Розділ 1: SELECT, FROM, WHERE

```python
# Імпортуємо потрібні бібліотеки
import pandas as pd
import sqlite3

# Підключаємося до бази даних
conn = sqlite3.connect("titanic.db")

# Якщо база не існує, створимо її з CSV
try:
    pd.read_sql("SELECT COUNT(*) FROM passengers", conn)
    print("База даних готова!")
except:
    titanic = pd.read_csv("Titanic Dataset.csv")
    titanic.to_sql("passengers", conn, if_exists="replace", index=False)
    print("Базу даних створено з CSV!")
```

```python
# Запит: SELECT всі стовпці, перші 5 рядків
query = "SELECT * FROM passengers LIMIT 5"
result = pd.read_sql(query, conn)
result
```

```python
# Запит: SELECT конкретні стовпці
query = "SELECT name, age FROM passengers LIMIT 10"
result = pd.read_sql(query, conn)
result
```

```python
# Запит: WHERE — фільтруємо тих хто вижив (survived = 1)
query = "SELECT name, age, survived FROM passengers WHERE survived = 1 LIMIT 10"
result = pd.read_sql(query, conn)
result
```

```python
# WHERE з оператором > — пасажири старші за 50
query = "SELECT name, age FROM passengers WHERE age > 50 LIMIT 10"
result = pd.read_sql(query, conn)
result
```

```python
# WHERE з текстовими значеннями в лапках — тільки жінки
query = "SELECT name, sex FROM passengers WHERE sex = 'female' LIMIT 10"
result = pd.read_sql(query, conn)
result
```

## Розділ 2: AND/OR/NOT, LIKE, BETWEEN, IN

```python
# AND — жінки які вижили
query = "SELECT name, sex, survived FROM passengers WHERE sex = 'female' AND survived = 1 LIMIT 10"
result = pd.read_sql(query, conn)
result
```

```python
# OR — пасажири першого АБО другого класу
query = "SELECT name, pclass FROM passengers WHERE pclass = 1 OR pclass = 2 LIMIT 10"
result = pd.read_sql(query, conn)
result
```

```python
# LIKE з % — імена що містять 'Mrs.'
query = "SELECT name FROM passengers WHERE name LIKE '%Mrs.%' LIMIT 10"
result = pd.read_sql(query, conn)
result
```

```python
# BETWEEN — пасажири віком від 20 до 40
query = "SELECT name, age FROM passengers WHERE age BETWEEN 20 AND 40 LIMIT 10"
result = pd.read_sql(query, conn)
result
```

```python
# IN — пасажири з Cherbourg або Queenstown
query = "SELECT name, embarked FROM passengers WHERE embarked IN ('C', 'Q') LIMIT 10"
result = pd.read_sql(query, conn)
result
```

## Розділ 3: ORDER BY та LIMIT

```python
# ORDER BY ASC — сортуємо за віком від наймолодшого
query = "SELECT name, age FROM passengers ORDER BY age ASC LIMIT 10"
result = pd.read_sql(query, conn)
result
```

```python
# ORDER BY DESC — сортуємо за ціною від найдорожчої
query = "SELECT name, fare FROM passengers ORDER BY fare DESC LIMIT 10"
result = pd.read_sql(query, conn)
result
```

```python
# Повний комбінований запит: SELECT + WHERE + ORDER BY + LIMIT
# Знаходимо жінок які вижили, сортуємо за ціною квитка
query = "SELECT name, age, fare FROM passengers WHERE sex = 'female' AND survived = 1 ORDER BY fare DESC LIMIT 10"
result = pd.read_sql(query, conn)
result
```
