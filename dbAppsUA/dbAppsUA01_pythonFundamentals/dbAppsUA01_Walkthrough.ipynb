## Розділ 2: Змінні та типи даних (Variables & Data Types)

**Змінна (variable)** — це контейнер, який зберігає значення. У Python є 4 основні типи даних:

1. **str** (string / рядок): Текстові дані — завжди в лапках. Приклад: `"Аліса"`
2. **int** (integer / ціле число): Цілі числа без десяткових. Приклад: `17`
3. **float** (число з комою): Числа з десятковими. Приклад: `3.75`
4. **bool** (boolean / логічний): Тільки True або False. Приклад: `True`

**Важливо:** У цьому курсі ми використовуємо **camelCase** для назв змінних.

```python
# Створюємо змінні різних типів даних
studentName = "Олена Петренко"    # str — текстові дані, завжди в лапках
studentAge = 17                   # int — ціле число, без десяткових
studentGPA = 3.75                 # float — число з десятковою крапкою
isEnrolled = True                 # bool — True або False

# Виводимо кожну змінну
print("Ім'я:", studentName)
print("Вік:", studentAge)
print("Середній бал:", studentGPA)
print("Зараховано:", isEnrolled)
```

```python
# Використовуємо type() щоб перевірити тип даних змінної
print("Тип studentName:", type(studentName))
print("Тип studentAge:", type(studentAge))
print("Тип studentGPA:", type(studentGPA))
print("Тип isEnrolled:", type(isEnrolled))
```

## Розділ 3: print() та коментарі (Comments)

**Функція print()** виводить текст або значення змінних на екран.

**Коментарі (comments)** — це нотатки в коді. Python їх ігнорує, але вони допомагають зрозуміти код. Починаються з `#`.

```python
# Різні способи використання print()

# Спосіб 1: Простий текст
print("Привіт, Застосування баз даних!")

# Спосіб 2: Друкуємо змінну
courseName = "Застосування баз даних"
print(courseName)

# Спосіб 3: Текст і змінні разом
courseCode = 145085
print("Курс:", courseName, "- Код:", courseCode)

# Спосіб 4: f-string — змінні прямо в тексті
print(f"Ласкаво просимо до {courseName} (Курс {courseCode})!")
```

## Розділ 4: Основи математики (Basic Math)

Python вміє рахувати! Ось арифметичні оператори:

- `+` Додавання (addition)
- `-` Віднімання (subtraction)
- `*` Множення (multiplication)
- `/` Ділення (division) — результат завжди float
- `**` Степінь (exponent)

```python
# Арифметичні оператори Python
print("Додавання:    5 + 3  =", 5 + 3)
print("Віднімання:   10 - 4 =", 10 - 4)
print("Множення:     3 * 7  =", 3 * 7)
print("Ділення:      15 / 4 =", 15 / 4)
print("Степінь:      2 ** 3 =", 2 ** 3)
```

```python
# Практичний приклад: Рахуємо вартість квитків у кіно
ticketPrice = 12.50    # Ціна одного квитка в доларах
numTickets = 4         # Кількість квитків

totalCost = ticketPrice * numTickets   # Множимо ціну на кількість

print(f"Ціна квитка: ${ticketPrice}")
print(f"Кількість квитків: {numTickets}")
print(f"Загальна вартість: ${totalCost}")
```

## Розділ 5: Імпорт pandas та завантаження даних

**pandas** — це бібліотека Python для роботи з даними. **DataFrame** — це як таблиця Excel: рядки та стовпці.

Ми будемо працювати з набором даних Titanic (Титанік).

```python
# Імпортуємо бібліотеку pandas
# Використовуємо псевдонім 'pd' для скорочення
import pandas as pd

print("pandas імпортовано успішно!")
print("Версія pandas:", pd.__version__)
```

```python
# Завантажуємо набір даних Titanic з CSV файлу
titanic = pd.read_csv("Titanic Dataset.csv")

print("Дані завантажено успішно!")
print(f"Набір даних має {titanic.shape[0]} рядків і {titanic.shape[1]} стовпців")
```

## Розділ 6: Дослідження даних (Exploring Data)

Перед аналізом даних потрібно їх **дослідити**. Основні методи:

- `.head()` — перші кілька рядків
- `.shape` — кількість рядків і стовпців
- `.info()` — детальна інформація про стовпці
- `.describe()` — статистика для числових стовпців
- `.columns` — список назв стовпців

```python
# Переглядаємо перші 5 рядків набору даних
titanic.head()
```

```python
# Перевіряємо розмір: (рядки, стовпці)
print("Розмір:", titanic.shape)
print(f"Набір має {titanic.shape[0]} пасажирів і {titanic.shape[1]} стовпців інформації")
```

```python
# Детальна інформація про кожен стовпець
titanic.info()
```

```python
# Статистика для числових стовпців
titanic.describe()
```
