**SQL Quick Reference:**
```sql
SELECT col1, col2 FROM table WHERE condition        -- pick columns, filter rows
WHERE col LIKE '%keyword%'                           -- partial text match (% = wildcard)
ORDER BY col DESC LIMIT 10                           -- sort + cap results
JOIN table2 ON t1.col = t2.col                       -- combine two tables
GROUP BY col                                         -- group for COUNT(), AVG()
IN ('val1', 'val2')                                  -- match any in list
IS NOT NULL                                          -- skip blank values
ROUND(AVG(col), 2)                                   -- round averages
COUNT(DISTINCT col)                                  -- count unique values
```
