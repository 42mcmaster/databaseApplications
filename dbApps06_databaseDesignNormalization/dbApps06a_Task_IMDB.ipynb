## Summary

Today you learned:
- **Redundancy** is storing the same data over and over (and why it's bad)
- **Primary keys** uniquely identify each row (like `tconst` and `nconst`)
- **Foreign keys** connect tables by referencing another table's PK
- The IMDb database uses these concepts to avoid hundreds of thousands of duplicate entries
