## Setup: Import Libraries and Connect

```python
import sqlite3
import pandas as pd

# Connect to NBA database
conn = sqlite3.connect('nba_5seasons.db')
print("✅ Connected to NBA database!")
```

---
## Task 1: Understanding Primary Keys

Examine the `players` table and verify its primary key.

### Task 1a: View table structure

```python
# Write a query to show the table structure for players
query = """
PRAGMA table_info(players);
"""

result = pd.read_sql(query, conn)
print(result)
```

**Question:** What column is the primary key? How can you tell?

**Answer:** `player_id` is the primary key. You can tell because the `pk` column shows `1` for player_id, indicating it's the primary key.

### Task 1b: Verify uniqueness

Write a query to verify that player_id has no duplicates.

```python
# Count total rows and unique player_id values
query = """
SELECT 
    COUNT(*) AS total_rows,
    COUNT(DISTINCT player_id) AS unique_player_ids
FROM players;
"""

result = pd.read_sql(query, conn)
print(result)
print("\n✅ If total_rows = unique_player_ids, every player_id is unique!")
```

## Task 2: Understanding Foreign Keys

Explore how the `team_game_stats` table connects to the `teams` table.

### Task 2a: Show games with team names

Write a query that joins `team_game_stats` to `teams` to show team names instead of just team_id.

```python
# Join team_game_stats to teams
query = """
SELECT 
    t.full_name AS team,
    tgs.pts AS points,
    tgs.reb AS rebounds
FROM team_game_stats tgs
JOIN teams t ON tgs.team_id = t.team_id
WHERE tgs.season = '2021-22'
LIMIT 10;
"""

result = pd.read_sql(query, conn)
print(result)
```

### Task 2b: Explain the foreign key

**Question:** What column in `team_game_stats` is the foreign key? What table does it reference?

**Answer:** The `team_id` column in `team_game_stats` is the foreign key. It references the `team_id` primary key in the `teams` table.

## Task 3: Understanding Data Redundancy

Calculate how much storage we save by NOT repeating team information.

### Task 3a: Count Cleveland Cavaliers games
How many games did the Cavaliers play in the 2021-22 season?

```python
# Count Cavaliers games in 2021-22 season
query = """
SELECT COUNT(*) AS cavaliers_games
FROM team_game_stats tgs
JOIN teams t ON tgs.team_id = t.team_id
WHERE t.full_name = 'Cleveland Cavaliers'
    AND tgs.season = '2021-22';
"""

result = pd.read_sql(query, conn)
print(result)
print("\nThe Cavaliers played 82 games in the 2021-22 season.")
```

### Task 3b: Calculate storage difference

Compare how many times team names are stored in our good design vs. a bad design.

```python
# Good design: Count rows in teams table
query_good = """
SELECT COUNT(*) AS team_names_stored
FROM teams;
"""

# Bad design: Count all games in 2021-22
query_bad = """
SELECT COUNT(*) AS team_names_if_repeated
FROM team_game_stats
WHERE season = '2021-22';
"""

good_count = pd.read_sql(query_good, conn).iloc[0, 0]
bad_count = pd.read_sql(query_bad, conn).iloc[0, 0]

print(f"Good design: {good_count} team names stored")
print(f"Bad design: {bad_count} team names stored")
print(f"We save {bad_count - good_count} duplicates!")
```

**Question:** If a team changed its name, how many rows would you need to update in each design?

**Answer:**
- Good design: **1 row** (just update the teams table)
- Bad design: **2,460 rows** (every game record for that team in one season, or tens of thousands across all seasons!)

## Task 4: Analyzing Table Relationships
Understand how players and teams connect through `player_season_stats`.

### Task 4a: Find players who changed teams

Write a query to find players who played for multiple teams in different seasons.

```python
# Find players with multiple teams across their career
query = """
SELECT 
    p.full_name AS player,
    COUNT(DISTINCT pss.team_id) AS num_teams
FROM player_season_stats pss
JOIN players p ON pss.player_id = p.player_id
GROUP BY p.player_id, p.full_name
HAVING COUNT(DISTINCT pss.team_id) > 1
ORDER BY num_teams DESC
LIMIT 10;
"""

result = pd.read_sql(query, conn)
print(result)
print("\nThese players have played for multiple teams!")
```

### Task 4b: Explain the relationship 

**Question:** Why can't we just put team_id directly in the players table?

**Answer:** Because players can play for different teams over their career. If we put team_id in the players table, we could only store ONE team per player. By using the player_season_stats table, we can track which team a player played for in each season. This creates a many-to-many relationship between players and teams.
