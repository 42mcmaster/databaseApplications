## Part 1: Examining Database Structure

Let's start by looking at what tables we have and their schemas.

### Step 1.1: List All Tables

````python
# Ask SQLite what tables exist in THIS file
tables = pd.read_sql("""
    SELECT name
    FROM sqlite_master
    WHERE type = 'table';
""", conn)

tables
````

### What we thought we had:
- From our previous work we knew we had 4 tables: teams, players, team_game_stats, player_season_stats:
  - **teams** = information about NBA teams
  - **players** = information about NBA players
  - **team_game_stats** = statistics for teams in individual games
  - **player_season_stats** = season statistics for players
---

### But Why are there extra tables?

SQLite shows **all tables in the database file**, not just the ones we’re using for this project.
The NBA tables (`teams`, `players`, `team_game_stats`, `player_season_stats`) are the ones we created intentionally.
The others (`books`, `members`, `checkouts`) are leftover from earlier examples, and `sqlite_sequence` is an internal SQLite system table.

➡️ These extra tables are **not errors** and can be safely ignored for our NBA work.


### Step 1.2: Examine teams Table Structure

````python
# Get schema for teams table
query = "PRAGMA table_info(teams)"
teams_schema = pd.read_sql(query, conn)
print("teams table schema:")
print(teams_schema)
````

````python
# Look at sample data
query = "SELECT * FROM teams LIMIT 5"
teams_sample = pd.read_sql(query, conn)
print("\nSample teams data:")
print(teams_sample)
````

**Analysis of teams table:**

**Primary Key:** `team_id` (INTEGER, unique identifier)
- ✅ Good choice: Surrogate key, never changes
- ✅ Better than using `full_name` which could theoretically change

**Attributes:**
- All attributes describe the team itself
- No repeating groups
- Each cell contains atomic values

**Normalization:**
- ✅ **1NF:** Atomic values, no repeating groups
- ✅ **3NF:** All attributes depend only on team_id (no transitive dependencies)

**Conclusion:** This table is well-designed and normalized!

### Step 1.3: Examine players Table Structure

````python
# Get schema for players table
query = "PRAGMA table_info(players)"
players_schema = pd.read_sql(query, conn)
print("players table schema:")
print(players_schema)
````

````python
# Look at sample data
query = "SELECT * FROM players LIMIT 5"
players_sample = pd.read_sql(query, conn)
print("\nSample players data:")
print(players_sample)
````

**Analysis of players table:**

**Primary Key:** `player_id` (INTEGER, unique identifier)
- ✅ Surrogate key, perfect for primary key

**Important Design Decision:**
- ❓ **Question:** Why is there NO `team_id` column here?
- ✅ **Answer:** Players change teams frequently!
  - Can't store current team because it changes
  - Historical data would be lost
  - Solution: Store team affiliation in `player_season_stats` table

**Normalization:**
- ✅ **1NF:** Atomic values
- ✅ **3NF:** All attributes depend only on player_id

**Conclusion:** Simple, clean, well-designed!

### Step 1.4: Examine team_game_stats Table Structure

````python
# Get schema for team_game_stats table
query = "PRAGMA table_info(team_game_stats)"
tgs_schema = pd.read_sql(query, conn)
print("team_game_stats table schema:")
print(tgs_schema)
````

````python
# Look at sample data
query = "SELECT * FROM team_game_stats LIMIT 5"
tgs_sample = pd.read_sql(query, conn)
print("\nSample team_game_stats data:")
print(tgs_sample)
````

**Analysis of team_game_stats table:**

**Primary Key:** `(game_id, team_id)` - **Composite key**
- ✅ Both columns needed for uniqueness
- Why? Each game has TWO teams (home and away)
- Same `game_id` appears twice (once per team)
- Together they uniquely identify each row

**Foreign Key:** `team_id` → teams(team_id)
- Links game stats to the team
- Maintains referential integrity

**Normalization:**
- ✅ **1NF:** Atomic values, no repeating groups
- ✅ **2NF:** All statistics depend on BOTH game_id AND team_id (no partial dependencies)
- ✅ **3NF:** No transitive dependencies

**Conclusion:** Properly normalized with composite key!

### Step 1.5: Examine player_season_stats Table Structure

````python
# Get schema for player_season_stats table
query = "PRAGMA table_info(player_season_stats)"
pss_schema = pd.read_sql(query, conn)
print("player_season_stats table schema:")
print(pss_schema)
````

````python
# Look at sample data
query = "SELECT * FROM player_season_stats LIMIT 5"
pss_sample = pd.read_sql(query, conn)
print("\nSample player_season_stats data:")
print(pss_sample)
````

**Analysis of player_season_stats table:**

**Primary Key:** `(player_id, team_id, season)` - **Composite key with 3 columns**
- ✅ All three needed for uniqueness
- Why?
  - Same player can play for different teams in same season (trades!)
  - Same player appears in multiple seasons
  - Together they uniquely identify each player-team-season combination

**Foreign Keys:**
- `player_id` → players(player_id)
- `team_id` → teams(team_id)

**Special Role:** **Junction Table**
- Creates many-to-many relationship:
  - One player plays for many teams (over career)
  - One team has many players (over seasons)
- Converts M:N to two 1:M relationships

**Normalization:**
- ✅ **2NF:** Stats depend on entire key (player + team + season)
- ✅ **3NF:** No transitive dependencies

**Conclusion:** Well-designed junction table!

## Part 2: Identifying Relationships

Let's verify the relationships between our tables.

### Relationship 1: teams ↔ team_game_stats (1:M)

````python
# How many games does each team have?
query = """
SELECT 
    t.full_name,
    COUNT(*) as game_count
FROM teams t
JOIN team_game_stats tgs ON t.team_id = tgs.team_id
WHERE tgs.season = '2021-22'
GROUP BY t.team_id, t.full_name
ORDER BY game_count DESC
LIMIT 5
"""

result = pd.read_sql(query, conn)
print("Games per team (2021-22 season):")
print(result)
````

**Analysis:**
- ✅ Each team has ~82 games (full NBA season)
- ✅ Confirms 1:M relationship: ONE team → MANY game stats
- Foreign key `team_id` in team_game_stats links to teams

### Relationship 2: teams ↔ player_season_stats (1:M)

````python
# How many player-seasons does each team have?
query = """
SELECT 
    t.full_name,
    COUNT(*) as player_seasons
FROM teams t
JOIN player_season_stats pss ON t.team_id = pss.team_id
WHERE pss.season = '2021-22'
GROUP BY t.team_id, t.full_name
ORDER BY player_seasons DESC
LIMIT 5
"""

result = pd.read_sql(query, conn)
print("Player-seasons per team (2021-22):")
print(result)
````

**Analysis:**
- ✅ Each team has multiple players in a season
- ✅ Confirms 1:M relationship: ONE team → MANY player-season records

### Relationship 3: players ↔ player_season_stats (1:M)

````python
# How many seasons does each player have?
query = """
SELECT 
    p.full_name,
    COUNT(*) as season_count
FROM players p
JOIN player_season_stats pss ON p.player_id = pss.player_id
GROUP BY p.player_id, p.full_name
ORDER BY season_count DESC
LIMIT 10
"""

result = pd.read_sql(query, conn)
print("Most seasons played (including team changes):")
print(result)
````

**Analysis:**
- ✅ Players can have multiple season records
- ✅ Confirms 1:M relationship: ONE player → MANY season records
- Players who changed teams will have multiple records per season!

### Relationship 4: players ↔ teams (M:N via junction table)

````python
# Find players who played for multiple teams in one season
query = """
SELECT 
    p.full_name,
    pss.season,
    COUNT(DISTINCT pss.team_id) as team_count,
    GROUP_CONCAT(t.abbreviation, ', ') as teams
FROM players p
JOIN player_season_stats pss ON p.player_id = pss.player_id
JOIN teams t ON pss.team_id = t.team_id
WHERE pss.season = '2021-22'
GROUP BY p.player_id, pss.season
HAVING COUNT(DISTINCT pss.team_id) > 1
LIMIT 10
"""

result = pd.read_sql(query, conn)
print("Players who changed teams in 2021-22:")
print(result)
````

**Analysis:**
- ✅ Some players played for multiple teams in same season (trades!)
- ✅ This demonstrates the M:N relationship:
  - One player can play for many teams
  - One team has many players
- ✅ player_season_stats acts as the junction table

**Why we need junction table:**
- Can't store team_id in players (which team?)
- Can't store player_id in teams (which players?)
- Solution: Separate table with BOTH foreign keys

## Part 3: Normalization Practice

Let's practice identifying and fixing normalization violations.

### Example 1: Violating 1NF (Multiple Values in Cell)

**BAD Design:**
```
Players_Bad:
player_id | name          | positions
1         | LeBron James  | SF, PF
2         | Stephen Curry | PG, SG
```

**Problem:** Multiple positions in one cell (not atomic)

**GOOD Design:**
```
Players:
player_id | name
1         | LeBron James
2         | Stephen Curry

Player_Positions:
player_id | position
1         | SF
1         | PF
2         | PG
2         | SG
```

````python
# Create the BAD table (for demonstration)
query = """
CREATE TABLE IF NOT EXISTS players_bad (
    player_id INTEGER PRIMARY KEY,
    name TEXT,
    positions TEXT
)
"""
conn.execute(query)

# Insert sample data
query = """
INSERT OR REPLACE INTO players_bad VALUES
    (1, 'LeBron James', 'SF, PF'),
    (2, 'Stephen Curry', 'PG, SG')
"""
conn.execute(query)
conn.commit()

# Show the problem
query = "SELECT * FROM players_bad"
print("BAD Design (violates 1NF):")
print(pd.read_sql(query, conn))
````

**Why this is bad:**
- Can't easily query "all players who play SF"
- Would need string parsing: `WHERE positions LIKE '%SF%'`
- Violates 1NF: cells are not atomic

````python
# Create the GOOD tables
query = """
CREATE TABLE IF NOT EXISTS players_good (
    player_id INTEGER PRIMARY KEY,
    name TEXT
)
"""
conn.execute(query)

query = """
CREATE TABLE IF NOT EXISTS player_positions (
    player_id INTEGER,
    position TEXT,
    PRIMARY KEY (player_id, position),
    FOREIGN KEY (player_id) REFERENCES players_good(player_id)
)
"""
conn.execute(query)

# Insert data
query = """
INSERT OR REPLACE INTO players_good VALUES
    (1, 'LeBron James'),
    (2, 'Stephen Curry')
"""
conn.execute(query)

query = """
INSERT OR REPLACE INTO player_positions VALUES
    (1, 'SF'),
    (1, 'PF'),
    (2, 'PG'),
    (2, 'SG')
"""
conn.execute(query)
conn.commit()

# Show the solution
print("GOOD Design (1NF):")
print("\nplayers_good:")
print(pd.read_sql("SELECT * FROM players_good", conn))
print("\nplayer_positions:")
print(pd.read_sql("SELECT * FROM player_positions", conn))
````

**Why this is better:**
- ✅ Each cell has one value (atomic)
- ✅ Easy to query: `WHERE position = 'SF'`
- ✅ Follows 1NF rules
- ✅ Each position is its own row

### Example 2: Violating 3NF (Transitive Dependency)

**BAD Design:**
```
Player_Stats_Bad:
player_id | player_name | team_id | team_name     | team_city
1         | LeBron      | 1610612747 | Lakers     | Los Angeles
2         | Curry       | 1610612744 | Warriors   | San Francisco
```

**Problem:** team_name and team_city depend on team_id, not player_id (transitive dependency)

````python
# Create BAD table
query = """
CREATE TABLE IF NOT EXISTS player_stats_bad (
    player_id INTEGER PRIMARY KEY,
    player_name TEXT,
    team_id INTEGER,
    team_name TEXT,
    team_city TEXT
)
"""
conn.execute(query)

# Insert sample data
query = """
INSERT OR REPLACE INTO player_stats_bad VALUES
    (1, 'LeBron James', 1610612747, 'Los Angeles Lakers', 'Los Angeles'),
    (2, 'Anthony Davis', 1610612747, 'Los Angeles Lakers', 'Los Angeles'),
    (3, 'Stephen Curry', 1610612744, 'Golden State Warriors', 'San Francisco')
"""
conn.execute(query)
conn.commit()

print("BAD Design (violates 3NF):")
print(pd.read_sql("SELECT * FROM player_stats_bad", conn))
````

**Problems with this design:**
1. Team information (name, city) repeated for each Lakers player
2. **Transitive dependency:** player_id → team_id → team_name, team_city
3. **Update anomaly:** If Lakers change city, must update ALL Lakers players
4. **Wasted space:** Same team info stored many times

**GOOD Design:**
```
Players:
player_id | player_name | team_id

Teams:
team_id | team_name | team_city
```

**Our actual NBA database already has this right:**
- players table (with team info in player_season_stats)
- teams table (separate)
- No redundant team data!

## Part 4: Design Verification Summary

### NBA Database Design Quality Check

**✅ teams table:**
- Primary Key: team_id (surrogate key)
- Normalization: 3NF
- No redundancy

**✅ players table:**
- Primary Key: player_id (surrogate key)
- Normalization: 3NF
- No team_id (correct - players change teams)

**✅ team_game_stats table:**
- Primary Key: (game_id, team_id) composite
- Foreign Key: team_id → teams
- Normalization: 2NF/3NF
- Properly handles 2 teams per game

**✅ player_season_stats table:**
- Primary Key: (player_id, team_id, season) composite
- Foreign Keys: player_id → players, team_id → teams
- Normalization: 2NF/3NF
- Junction table for players ↔ teams M:N
- Handles mid-season trades

**Overall Assessment:**
- ✅ Well-designed database
- ✅ Properly normalized
- ✅ Appropriate use of surrogate keys
- ✅ Foreign keys maintain referential integrity
- ✅ Handles real-world scenarios (trades, multiple seasons)

## Key Takeaways

From this walkthrough, you learned:

1. **How to analyze database structure:**
   - Use PRAGMA table_info() to see schema
   - Identify primary and foreign keys
   - Verify relationships between tables

2. **Normalization principles:**
   - 1NF: Atomic values, no repeating groups
   - 2NF: No partial dependencies (composite keys)
   - 3NF: No transitive dependencies

3. **Good design patterns:**
   - Use surrogate keys for primary keys
   - Use foreign keys to link tables
   - Create junction tables for M:N relationships
   - Separate concerns into appropriate tables

4. **Our NBA database:**
   - Well-designed and normalized
   - Handles complex real-world scenarios
   - Uses composite keys where appropriate
   - Maintains data integrity with foreign keys
