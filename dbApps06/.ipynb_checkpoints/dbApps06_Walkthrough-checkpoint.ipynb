## Part 2: Single-Table vs Multi-Table Design

Let's see what happens if we try to put everything in ONE table.

### Step 2.1: Create a Denormalized (Single-Table) View

```python
# Join all tables together to simulate single-table design
query = """
SELECT 
    p.full_name,
    t.full_name,
    t.city,
    ps.season,
    ps.gp as games_played,
    ps.pts as points_per_game,
    ps.reb as rebounds_per_game,
    ps.ast as assists_per_game
FROM player_season_stats ps
JOIN players p ON ps.player_id = p.player_id
JOIN teams t ON ps.team_id = t.team_id
WHERE ps.season = '2023-24'
LIMIT 10
"""

denormalized = pd.read_sql(query, conn)
print("This is what a single-table design would look like:")
display(denormalized)
```

### Step 2.2: Identify the Problems

**Problem 1: Data Redundancy**

```python
# How many times is "Lakers" repeated?
query = """
SELECT 
    t.full_name,
    COUNT(*) as times_repeated
FROM player_season_stats ps
JOIN teams t ON ps.team_id = t.team_id
WHERE t.full_name = 'Los Angeles Lakers'
  AND ps.season = '2023-24'
"""

redundancy = pd.read_sql(query, conn)
print("Team name 'Los Angeles Lakers' appears this many times:")
display(redundancy)

print("\n💡 In a single table, we'd store 'Los Angeles Lakers' 15+ times!")
print("In our multi-table design, we store it ONCE in the teams table.")
```

**Problem 2: Update Anomaly**

Imagine the Lakers move to a new city (hypothetically).

```python
# Multi-table approach (GOOD)
print("Multi-table approach: Update ONE row")
print("SQL: UPDATE teams SET city = 'San Diego' WHERE full_name = 'Los Angeles Lakers'")
print("✅ Updates 1 row")
print("✅ Guaranteed consistency")
print("✅ Fast operation")

print("\n" + "="*50)

# Single-table approach (BAD)
print("Single-table approach: Update MANY rows")
print("SQL: UPDATE all_data SET city = 'San Diego' WHERE full_name = 'Los Angeles Lakers'")
print("❌ Updates 100+ rows (every player, every game)")
print("❌ Risk of missing some rows")
print("❌ Slow on large tables")
print("❌ Risk of typos causing inconsistency")
```

**Discussion Question:**
> What if you update 99 rows but miss 1 row? Now your database has inconsistent data!

## Part 3: Primary Keys vs Foreign Keys

Let's understand the difference between primary keys and foreign keys.

### Step 3.1: Primary Keys - Unique Identifiers

```python
# Primary key: Uniquely identifies each team
query = """
SELECT 
    team_id,          -- PRIMARY KEY
    full_name,
    city
FROM teams
LIMIT 5
"""

teams = pd.read_sql(query, conn)
print("Teams table - team_id is the PRIMARY KEY:")
display(teams)

print("\n💡 Each team_id is unique and identifies exactly one team.")
```

```python
# Check: Are all team_ids unique?
query = """
SELECT 
    team_id,
    COUNT(*) as count
FROM teams
GROUP BY team_id
HAVING COUNT(*) > 1
"""

duplicates = pd.read_sql(query, conn)

if len(duplicates) == 0:
    print("✅ All team_ids are unique - it's a valid primary key!")
else:
    print("❌ Found duplicate team_ids:")
    display(duplicates)
```

### Step 3.2: Foreign Keys - Creating Relationships

```python
# Foreign key: Links players to teams through player_season_stats
query = """
SELECT 
    p.player_id,      -- PRIMARY KEY in players table
    p.full_name,
    ps.team_id,       -- FOREIGN KEY in junction table (references teams.team_id)
    ps.season,
    t.full_name as team_name       -- Information from related table
FROM players p
JOIN player_season_stats ps ON p.player_id = ps.player_id
JOIN teams t ON ps.team_id = t.team_id
WHERE ps.season = '2023-24'
LIMIT 5
"""

players = pd.read_sql(query, conn)
print("Player-Team relationship through player_season_stats:")
display(players)

print("\n💡 Players don't have team_id directly - the relationship goes through player_season_stats!")
print("   This is because players can change teams (trades).")
```

### Step 3.3: Referential Integrity Demo

Foreign keys enforce **referential integrity** - you can't reference non-existent records.

```python
# Check if all player_season_stats team_ids reference valid teams
query = """
SELECT 
    ps.player_id,
    ps.team_id,
    ps.season
FROM player_season_stats ps
LEFT JOIN teams t ON ps.team_id = t.team_id
WHERE t.team_id IS NULL
LIMIT 5
"""

orphaned = pd.read_sql(query, conn)

if len(orphaned) == 0:
    print("✅ All player_season_stats reference valid teams - referential integrity maintained!")
else:
    print("❌ Found player-season records with invalid team_ids:")
    display(orphaned)

print("\n💡 Foreign keys prevent you from inserting invalid team_ids into player_season_stats.")
```

## Part 4: Normal Forms - Violations and Fixes

Let's see examples of normal form violations and how to fix them.

### Example 1: First Normal Form (1NF) Violation

**1NF Rule:** All values must be atomic (single value, no lists).

```python
# Simulating a 1NF violation
bad_design = pd.DataFrame({
    'full_name': ['LeBron James', 'Kevin Durant'],
    'positions': ['SF, PF', 'SF, PF, SG'],  # ❌ Multiple values in one cell!
    'teams_played_for': ['Cavs, Heat, Lakers', 'Thunder, Warriors, Nets']  # ❌ List in cell!
})

print("❌ NOT in 1NF - Multiple values in single cells:")
display(bad_design)

print("\nProblems:")
print("- Can't easily query 'Find all players who play PF'")
print("- Can't count how many positions each player has")
print("- Hard to sort or filter")
```

```python
# Fixed: Separate rows for each value
good_design = pd.DataFrame({
    'full_name': ['LeBron James', 'LeBron James', 'Kevin Durant', 'Kevin Durant', 'Kevin Durant'],
    'position': ['SF', 'PF', 'SF', 'PF', 'SG']
})

print("✅ Fixed - Now in 1NF (atomic values):")
display(good_design)

print("\nBenefits:")
print("- ✅ Easy to query: SELECT * FROM positions WHERE position = 'PF'")
print("- ✅ Easy to count: COUNT(*) GROUP BY position")
print("- ✅ Proper database structure")
```

### Example 2: Second Normal Form (2NF) Violation

**2NF Rule:** No partial dependencies (all non-key columns must depend on the ENTIRE primary key).

**Only applies when you have a composite primary key!**

```python
# Simulating a 2NF violation
bad_design = pd.DataFrame({
    'player_id': [1, 1, 2, 2],
    'game_id': [100, 101, 100, 101],
    'pts': [25, 30, 22, 28],
    'full_name': ['LeBron', 'LeBron', 'AD', 'AD'],  # ❌ Depends only on player_id
    'full_name': ['Lakers', 'Lakers', 'Lakers', 'Lakers']  # ❌ Depends only on player_id
})

print("❌ NOT in 2NF - full_name and full_name depend only on player_id, not on (player_id, game_id):")
display(bad_design)

print("\nProblems:")
print("- Redundancy: 'LeBron' and 'Lakers' repeated in every game")
print("- Update anomaly: Change player name → must update many rows")
```

```python
# Fixed: Split into two tables
players_table = pd.DataFrame({
    'player_id': [1, 2],
    'full_name': ['LeBron', 'AD'],
    'full_name': ['Lakers', 'Lakers']
})

game_stats = pd.DataFrame({
    'player_id': [1, 1, 2, 2],
    'game_id': [100, 101, 100, 101],
    'pts': [25, 30, 22, 28]
})

print("✅ Fixed - Now in 2NF:")
print("\nPlayers table (player info):")
display(players_table)

print("\nGame stats table (game-specific data):")
display(game_stats)

print("\nBenefits:")
print("- ✅ Player info stored once")
print("- ✅ Update player name in one place")
```

### Example 3: Third Normal Form (3NF) Violation

**3NF Rule:** No transitive dependencies (non-key columns shouldn't depend on other non-key columns).

```python
# Simulating a 3NF violation
bad_design = pd.DataFrame({
    'player_id': [1, 2, 3],
    'full_name': ['LeBron', 'AD', 'Steph'],
    'team_id': [10, 10, 15],
    'full_name': ['Lakers', 'Lakers', 'Warriors'],  # ❌ Depends on team_id, not player_id
    'team_city': ['Los Angeles', 'Los Angeles', 'San Francisco']  # ❌ Depends on team_id
})

print("❌ NOT in 3NF - full_name and team_city depend on team_id (transitive dependency):")
display(bad_design)

print("\nProblems:")
print("- Redundancy: 'Lakers' and 'Los Angeles' repeated for every Lakers player")
print("- Update anomaly: Lakers move to new city → update all player rows")
```

```python
# Fixed: Split into players and teams tables
players_table = pd.DataFrame({
    'player_id': [1, 2, 3],
    'full_name': ['LeBron', 'AD', 'Steph'],
    'team_id': [10, 10, 15]  # Foreign key
})

teams_table = pd.DataFrame({
    'team_id': [10, 15],
    'full_name': ['Lakers', 'Warriors'],
    'team_city': ['Los Angeles', 'San Francisco']
})

print("✅ Fixed - Now in 3NF:")
print("\nPlayers table:")
display(players_table)

print("\nTeams table:")
display(teams_table)

print("\nBenefits:")
print("- ✅ Team info stored once")
print("- ✅ Update team city in one place")
print("- ✅ This is exactly how the NBA database is designed!")
```

### Example 4: Calculated Fields (3NF Violation)

```python
# Simulating storing calculated fields
bad_design = pd.DataFrame({
    'player_id': [1, 2],
    'fgm': [8, 10],  # Field goals made
    'fga': [15, 18],  # Field goals attempted
    'fg_pct': [0.533, 0.556]  # ❌ Calculated field! Can compute from fgm/fga
})

print("❌ Storing calculated fields (violates 3NF):")
display(bad_design)

print("\nProblems:")
print("- Risk: fg_pct might not match actual calculation")
print("- Update anomaly: Change fgm or fga → must also update fg_pct")
```

```python
# Fixed: Calculate when needed
good_design = pd.DataFrame({
    'player_id': [1, 2],
    'fgm': [8, 10],
    'fga': [15, 18]
})

# Calculate field goal percentage
good_design['fg_pct'] = (good_design['fgm'] / good_design['fga']).round(3)

print("✅ Calculate on demand:")
display(good_design)

print("\nBenefits:")
print("- ✅ Always accurate (calculated from source data)")
print("- ✅ No risk of inconsistency")

print("\nSQL equivalent:")
print("SELECT player_id, fgm, fga,")
print("       ROUND(CAST(fgm AS FLOAT) / fga, 3) as fg_pct")
print("FROM player_stats")
```

## Part 5: Design Exercise - Library Database

Let's design a simple library database together following normalization principles.

### Step 5.1: Identify Entities

**What are the main "things" in a library system?**

1. **Books** - The items in the library
2. **Members** - People who can check out books
3. **Checkouts** - Records of who borrowed what and when

**Each becomes a table!**

### Step 5.2: Define Attributes

**Books:**
- book_id (PRIMARY KEY)
- title
- author
- isbn
- publication_year

**Members:**
- member_id (PRIMARY KEY)
- first_name
- last_name
- email
- join_date

**Checkouts:**
- checkout_id (PRIMARY KEY)
- book_id (FOREIGN KEY → books)
- member_id (FOREIGN KEY → members)
- checkout_date
- due_date
- return_date

### Step 5.3: Create the Tables

```python
# Create books table
conn.execute('''
CREATE TABLE IF NOT EXISTS books (
    book_id INTEGER PRIMARY KEY AUTOINCREMENT,
    title TEXT NOT NULL,
    author TEXT NOT NULL,
    isbn TEXT UNIQUE,
    publication_year INTEGER
)
''')

# Create members table
conn.execute('''
CREATE TABLE IF NOT EXISTS members (
    member_id INTEGER PRIMARY KEY AUTOINCREMENT,
    first_name TEXT NOT NULL,
    last_name TEXT NOT NULL,
    email TEXT UNIQUE,
    join_date TEXT DEFAULT CURRENT_DATE
)
''')

# Create checkouts table with foreign keys
conn.execute('''
CREATE TABLE IF NOT EXISTS checkouts (
    checkout_id INTEGER PRIMARY KEY AUTOINCREMENT,
    book_id INTEGER NOT NULL,
    member_id INTEGER NOT NULL,
    checkout_date TEXT NOT NULL,
    due_date TEXT NOT NULL,
    return_date TEXT,
    FOREIGN KEY (book_id) REFERENCES books(book_id),
    FOREIGN KEY (member_id) REFERENCES members(member_id)
)
''')

conn.commit()
print("✅ Library database tables created!")
```

### Step 5.4: Add Sample Data

```python
# Add some books
conn.execute('''
INSERT INTO books (title, author, isbn, publication_year) VALUES
    ('The Great Gatsby', 'F. Scott Fitzgerald', '978-0743273565', 1925),
    ('1984', 'George Orwell', '978-0451524935', 1949),
    ('To Kill a Mockingbird', 'Harper Lee', '978-0060935467', 1960)
''')

# Add some members
conn.execute('''
INSERT INTO members (first_name, last_name, email) VALUES
    ('Alice', 'Johnson', 'alice@email.com'),
    ('Bob', 'Smith', 'bob@email.com'),
    ('Carol', 'Williams', 'carol@email.com')
''')

# Add some checkouts
conn.execute('''
INSERT INTO checkouts (book_id, member_id, checkout_date, due_date, return_date) VALUES
    (1, 1, '2024-01-15', '2024-02-15', '2024-02-10'),
    (2, 2, '2024-02-01', '2024-03-01', NULL),
    (3, 1, '2024-02-10', '2024-03-10', NULL)
''')

conn.commit()
print("✅ Sample data added!")
```

### Step 5.5: Query the Library Database

```python
# Query: Which books are currently checked out?
query = """
SELECT 
    b.title,
    b.author,
    m.first_name || ' ' || m.last_name as member_name,
    c.checkout_date,
    c.due_date
FROM checkouts c
JOIN books b ON c.book_id = b.book_id
JOIN members m ON c.member_id = m.member_id
WHERE c.return_date IS NULL
"""

checked_out = pd.read_sql(query, conn)
print("Books currently checked out:")
display(checked_out)
```

```python
# Query: What's Alice's checkout history?
query = """
SELECT 
    b.title,
    c.checkout_date,
    c.return_date,
    CASE 
        WHEN c.return_date IS NULL THEN 'Currently checked out'
        ELSE 'Returned'
    END as status
FROM checkouts c
JOIN books b ON c.book_id = b.book_id
JOIN members m ON c.member_id = m.member_id
WHERE m.first_name = 'Alice' AND m.last_name = 'Johnson'
"""

alice_history = pd.read_sql(query, conn)
print("Alice's checkout history:")
display(alice_history)
```

### Step 5.6: Verify Referential Integrity

```python
# Try to add a checkout for a non-existent book
try:
    conn.execute('''
    INSERT INTO checkouts (book_id, member_id, checkout_date, due_date)
    VALUES (999, 1, '2024-03-01', '2024-04-01')
    ''')
    print("❌ Uh oh, it allowed an invalid book_id!")
except sqlite3.IntegrityError as e:
    print("✅ Foreign key constraint prevented invalid data!")
    print(f"Error: {e}")
```

**Discussion:**
- Why is this library database design in 3NF?
- What problems would a single-table design have?
- What queries would be difficult without proper normalization?

## Summary & Key Takeaways

**What we learned:**

1. **Single-table designs have problems:**
   - Data redundancy
   - Update anomalies
   - Insertion anomalies
   - Deletion anomalies

2. **Normalization solves these problems:**
   - **1NF:** Atomic values, no repeating groups
   - **2NF:** No partial dependencies
   - **3NF:** No transitive dependencies

3. **Keys are crucial:**
   - **Primary Key:** Uniquely identifies each row
   - **Foreign Key:** Creates relationships between tables
   - **Referential Integrity:** Foreign keys must reference valid primary keys

4. **Database design process:**
   - Identify entities
   - Define attributes
   - Determine relationships
   - Create ERD
   - Implement in SQL
