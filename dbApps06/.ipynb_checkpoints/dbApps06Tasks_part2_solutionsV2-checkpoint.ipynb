## Task 2: Spot the Bad Design 

We're going to create a poorly designed table, then identify all its problems.

```python
# Create a BAD table design (run this cell as-is)
query = """
CREATE TEMP TABLE bad_player_data AS
SELECT 
    p.full_name AS player_name,
    pss.season,
    t.full_name AS team_name,
    t.city AS team_city,
    t.state AS team_state,
    pss.gp AS games_played,
    pss.pts AS points_per_game
FROM player_season_stats pss
JOIN players p ON pss.player_id = p.player_id
JOIN teams t ON pss.team_id = t.team_id
WHERE pss.season = '2021-22'
LIMIT 20;
"""

conn.execute(query)
print("✅ Created bad_player_data table")

# Look at the table
result = pd.read_sql("SELECT * FROM bad_player_data LIMIT 10", conn)
print("\nBAD TABLE DESIGN:")
print(result)
```

### Task 2a: Find duplicate team data

How many times does "Cleveland" appear in this table?

```python
# Count how many times Cleveland appears
query = """
SELECT COUNT(*) AS cleveland_count
FROM bad_player_data
WHERE team_city = 'Cleveland';
"""

result = pd.read_sql(query, conn)
print(result)
print("\n'Cleveland' is stored multiple times - once per player!")
```

### Task 2b: Identify all the problems

**Answer - Problems with this table design:**

1. **Data redundancy** - Team name, city, and state are repeated for every player on that team. If a team has 15 players, this info is stored 15 times!

2. **Wasted storage** - Storing "Cleveland Cavaliers", "Cleveland", and "Ohio" over and over wastes disk space.

3. **Update anomalies** - If Cleveland changes its team name, you'd have to update every single player record instead of just one row in a teams table.

4. **Inconsistency risk** - Someone could misspell "Cleveland" in one row, creating bad data.

5. **No primary key** - This table doesn't have a unique identifier, so you can't easily reference specific rows.

**How to fix this design:**

Split it into three separate tables:
- **players table**: player_id (PK), player_name
- **teams table**: team_id (PK), team_name, city, state
- **player_season_stats table**: player_id (FK), team_id (FK), season, games_played, points_per_game

This way, team info is stored once, and player stats just reference the team using team_id.

## Task 3: Practice Adding Data Correctly 

Let's practice inserting new records with proper foreign key relationships.

### Task 3a: Find the Cavaliers team_id 

```python
# Find the team_id for Cleveland Cavaliers
query = """
SELECT team_id, full_name
FROM teams
WHERE full_name = 'Cleveland Cavaliers';
"""

result = pd.read_sql(query, conn)
print(result)
print("\nThe Cavaliers team_id is: 1610612739")
```

### Task 3b: Explain the insertion process 

**Scenario:** You want to add a new player named "John Smith" who plays for the Cavaliers.

**Correct order:**
1. **Step 3** - Look up the Cavaliers team_id from the teams table (1610612739)
2. **Step 2** - Insert "John Smith" into the players table and get the new player_id
3. **Step 1** - Insert player stats into player_season_stats using both the new player_id and the team_id

**Why is this order important?**

Because of **referential integrity**! You can't insert a foreign key value that doesn't exist in the parent table.

- The team_id (1610612739) must already exist in the teams table before we can use it
- The player_id must exist in the players table before we can reference it in player_season_stats
- If we tried to insert stats with a player_id that doesn't exist, the database would reject it (or should, if foreign key constraints are enabled)

This is one of the key benefits of good database design - the database itself helps prevent bad data!

## Bonus Task: Design Improvement Ideas 

**Sample Answer:**

**What I would add:** Player positions (PG, SG, SF, PF, C)

**Where to add it:** Create a new column in the `player_season_stats` table called `position`

**Why this would be useful:** 
- Players sometimes change positions over their career or on different teams
- You could analyze stats by position (average points for point guards vs centers)
- Coaches and analysts often want to compare players at the same position
- It shouldn't go in the players table because a player's position can change between seasons or teams

**Other good answers:**
- Coach names (coaches table with team_id and season)
- Arena names (could add to teams table since most teams have one arena)
- Playoff vs regular season flag (add to team_game_stats)
- Player height/weight (add to players table since these don't change much)
- Conference/division (add to teams table)
