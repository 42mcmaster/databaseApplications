## Task 2: Spot the Bad Design

We're going to create a poorly designed table, then identify all its problems.

```python
# Create a BAD table design (run this cell as-is)
query = """
CREATE TEMP TABLE bad_player_data AS
SELECT 
    p.full_name AS player_name,
    pss.season,
    t.full_name AS team_name,
    t.city AS team_city,
    t.state AS team_state,
    pss.gp AS games_played,
    pss.pts AS points_per_game
FROM player_season_stats pss
JOIN players p ON pss.player_id = p.player_id
JOIN teams t ON pss.team_id = t.team_id
WHERE pss.season = '2021-22'
LIMIT 20;
"""

conn.execute(query)
print("✅ Created bad_player_data table")

# Look at the table
result = pd.read_sql("SELECT * FROM bad_player_data LIMIT 10", conn)
print("\nBAD TABLE DESIGN:")
print(result)
```
